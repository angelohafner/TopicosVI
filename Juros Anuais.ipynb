# Campos Vetoriais

Imaginemos a EDO do saldo banc√°rio

$$ \frac{d}{dt}S(t) = k \, S(t) $$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
k = 0.1
```

```python
tempo = np.linspace(0, 50, 10)
saldo = np.linspace(0, 400, 10)
```

```python
TEMPO, SALDO = np.meshgrid(tempo, saldo)
INCLINACOES = k * SALDO
ANGULOS = np.arctan(INCLINACOES)
COMP_X = np.cos(ANGULOS)
COMP_Y = np.sin(ANGULOS)
```

```python
plt.figure(dpi=100, figsize=(10,5))
plt.quiver(TEMPO, SALDO, COMP_X, COMP_Y, angles='xy')
for s0 in range(0, 10, 1):
    plt.plot(tempo, s0*np.exp(k*tempo), alpha=0.5, lw=5)
plt.xlabel('$t$ [anos]')
plt.ylabel('$s(t)$ [mil reais]')
plt.ylim(-10, 410)

plt.show()
```
