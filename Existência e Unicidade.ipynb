```python
import sympy as sp
import matplotlib.pyplot as plt
import numpy as np

y = sp.Function("y")
t = sp.Symbol('t')
```

```python
D1 = sp.Derivative(y(t), t)
D1
```

```python
t
```

```python
ode = sp.Eq( t*D1 + 2*y(t)  , 4*t**2 )
ode
```

```python
ySol = sp.dsolve( ode , ics={y(0): 0} ) 
ySol
```

```python

yNum = sp.lambdify(t,ySol.args[1])
```

```python
tt = np.linspace(-3,3,1000)
ode = sp.Eq( t*D1 + 2*y(t)  , 4*t**2 )

plt.figure(figsize=(4,3),dpi=300)
plt.ylim([-2,6])
plt.grid()
for y0 in range(-4,5,2):
  ySol = sp.dsolve( ode , ics={y(1): y0} ) 
  yNum = sp.lambdify(t,ySol.args[1])
  plt.plot(tt,yNum(tt),label=y0)
  
T = np.linspace(-3,3,10)
Y = np.linspace(-2,6,10)
T,Y = np.meshgrid(T,Y)
deriv = (4*T**2 - 2*Y) / T
ang = np.arctan(Y/T)
u = np.cos(ang)
v = np.sin(ang)
plt.quiver(T,Y,u,v,angles='xy')
```
