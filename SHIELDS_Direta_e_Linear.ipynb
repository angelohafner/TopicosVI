```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
x = sp.symbols('x')
```

```python
f = sp.Function('f')(x)
d = 1 + 2*sp.exp(2*x)
expr = sp.Derivative(f,x) - d
expr
```

```python
#eq = sp.Eq(f_,1+2*sp.exp(2*x))
#eq
```

```python
fSol = sp.dsolve(expr,f,ics={f.subs(x,0):0})
fSol
```

```python
fSym = fSol.args[1]
fSym
```

```python
fNum = sp.lambdify(x,fSym)
type(fNum)
```

```python
f = plt.figure()
f.set_figwidth(6)
f.set_figheight(8)

xx = np.linspace(-2,2,11);
yy = np.linspace(-10,10,11);
xx,yy = np.meshgrid(xx,yy)

```

```python
dNum = sp.lambdify(x,d)
dd = dNum(xx)
aa = np.arctan(dd)
u = np.cos(aa)
v = np.sin(aa)

plt.figure(figsize=(4,5),dpi=300)
plt.quiver(xx,yy,u,v,label='$dy/dx$',angles='xy')

xx = np.linspace(-2,2);
yy = fNum(xx)
plt.plot(xx,yy,label='$y$, $y(0)=0$',lw=3)
plt.grid()
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.ylim([-10,10])
plt.legend()
```
