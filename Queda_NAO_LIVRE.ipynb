# Exemplo de queda não livre

$$a = v' = g - \frac{k}{m}v$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
g = 10
k = 20
m = 80
vf = m*g/k
```

### Definindo os tempos e as velocidades nas quais se deseja save a inclinação

```python
nnn = 11
tempo = np.linspace(0,20,nnn)
velocidade = np.linspace(0.8*vf,1.2*vf,nnn)
```

```python
T,V = np.meshgrid(tempo,velocidade)
a = g - k/m*V
angulo = np.arctan(a)
dx = np.cos(angulo)
dy = np.sin(angulo)
```

```python
plt.quiver(T,V,dx,dy,angles='xy')
```

### Definição da EDO


$$a = v' = g - \frac{k}{m}v$$

```python
def edi(v,t,k,m,g):
    a = g - k/m*v
    return a
```

```python
import scipy.integrate as spi
ttt = np.linspace(0,20,100)
```

```python
plt.figure(figsize=(16/2,9/2),dpi=100)
plt.quiver(T,V,dx,dy,angles='xy',label="$a$")
for v0 in range(32,49,4):
    sol = spi.odeint(edi,v0,ttt,args=(k,m,g))
    vvv = "$v_0=$"+str(v0)
    plt.plot(ttt,sol,label=vvv,lw=3)
plt.legend(loc=1)
plt.xlabel("Tempo [s]")   
plt.ylabel("Velocidade [m/s]")    
plt.show()
```
