```python

import numpy as np
import matplotlib.pyplot as plt
```

```python
def edo(y,t,x,k):
  return -r*(1-y/x)*(1-y/k)*y
```

```python
tLim = 6
t = np.linspace(0,tLim,16)
y = np.linspace(0,15,16)
sc = 15/tLim

T,Y = np.meshgrid(t,y)
r = 1
k = 10
x = 4
dy_dt = -r*(1-Y/x)*(1-Y/k)*Y
ang = np.arctan(dy_dt)
u = sc*np.cos(ang)
v = np.sin(ang)
ang2 = np.arctan(v/u)
u = np.cos(ang2)
v = np.sin(ang2)

plt.figure(figsize=(8,8),dpi=150)
plt.quiver(T,Y,u,v,label='$dy/dt$')
plt.xlabel('Tempo')
plt.ylabel('População')

t = np.linspace(0,tLim,100)
from scipy.integrate import odeint
for y0 in range(0,16,1):
  sol = odeint(edo,y0,t,args=(x, k))
  plt.plot(t,sol,label=y0,lw=3)
plt.legend()
plt.grid()
```
