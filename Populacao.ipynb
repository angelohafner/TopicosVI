```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
tLim = 6
t = np.linspace(0,tLim,16)
y = np.linspace(0,15,16)
sc = 15/tLim
T,Y = np.meshgrid(t,y)
r = 1
k = 10

dy_dt = r*(1-Y/k)*Y
ang = np.arctan(dy_dt)
u = sc*np.cos(ang)
v = np.sin(ang)
ang2 = np.arctan(v/u)
u = np.cos(ang2)
v = np.sin(ang2)

plt.figure(figsize=(8,8),dpi=150)
plt.quiver(T,Y,u,v,label='$dy/dt$')
plt.xlabel('Tempo')
plt.ylabel('População')

y0 = 0
while( y0<=15):
  den = y0 + (k-y0)*np.exp(-r*t)
  popul = y0*k / den
  y0 = y0 + 1
  plt.plot(t,popul,lw=3,label=y0)
plt.grid()
plt.legend()
plt.show()
```
