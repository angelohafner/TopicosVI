```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
ggg = 9.8
gamma = 0.02
mmm = 10
ttt = np.linspace(0, 2000, 21)
vvv = np.linspace(0, 6000, 21)

TTT, VVV = np.meshgrid(ttt, vvv)
```

$$\frac{dv}{dt} = g - \frac{\gamma}{m} v$$

```python
aceleracao = ggg - gamma/mmm*VVV 
angulo = np.arctan(aceleracao)
comp_x = np.cos(angulo)
comp_y = np.sin(angulo)
plt.quiver(ttt, vvv, comp_x, comp_y, angles='xy')
plt.show()
```

```python

veloc = -mmm*ggg/gamma*np.exp((-gamma/mmm)*ttt) + mmm*ggg/gamma
plt.plot(ttt, veloc)
plt.quiver(ttt, vvv, comp_x, comp_y, angles='xy')
plt.show()
```
