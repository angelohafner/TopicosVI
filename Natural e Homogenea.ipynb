```python
import sympy as sp
y = sp.Function("y")
t = sp.Symbol('t')
```

```python
D1 = sp.Derivative(y(t), t)
D1
```

```python
D2 = sp.Derivative(y(t), t,t)
D2
```

```python
ode = sp.Eq( D2 - 3*D1 -4*y(t), 3*sp.exp(2*t)  )
ode
```

```python
solHomogenea = sp.dsolve( sp.Eq( D2 - 3*D1 -4*y(t),0) ,y(t) ) 
solHomogenea
```

```python
solTotal = sp.dsolve( ode,y(t),ics={y(0): 0, y(t).diff(t).subs(t, 0): 0})
solTotal
```

```python
f = solTotal.args[1]
yForcado = f.args[2]
yNatural = f.args[0] + f.args[1]
yNatural
```
