```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 4*np.pi, 2000)
y = np.sin(x)
```

```python
fig = plt.figure(dpi=100)
g1 = plt.plot(x, y, label="$\\sin \\,   x$")
g2 = plt.plot(x, y**2, label="$\\sin^2 \\, x$")
plt.legend()
plt.grid(ls='--', alpha=0.5)
plt.xlabel('$\\frac{T}{T_a}$')
plt.ylabel('Variável')
plt.title("Título do Gráfico")
plt.show()
```
