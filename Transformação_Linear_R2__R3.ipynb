# Transformação Linear

Seja a transformação linear

\begin{eqnarray*}
T:\mathbb{R}^2 &\longrightarrow& \mathbb{R}^3\\
(x,y) &\mapsto& (2x, 3y, x - 2y)
\end{eqnarray*}

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-4,4,11)
y = np.linspace(-4,4,11)
X, Y = np.meshgrid(x,y)
```

```python
X_transf = 2*X
Y_transf = 3*Y
Z_transf = X - 2*Y

xMax = np.array( np.max(X_transf) )
yMax = np.array( np.max(Z_transf) )
zMax = np.array( np.max(Z_transf) )
maximo = np.max([xMax,yMax,zMax])

x = np.linspace(-maximo,maximo,11)
y = np.linspace(-maximo,maximo,11)
X, Y = np.meshgrid(x,y)

```

```python
fig = plt.figure(figsize=(8,5),dpi=100)
ax1 = plt.axes(projection='3d')

surf = ax1.plot_surface(X,Y,0*X, alpha=0.5, edgecolors = 'blue', label='$<x,\, y>$')
surf._facecolors2d=surf._facecolors3d 
surf._edgecolors2d=surf._edgecolors3d
surf = ax1.plot_surface(X_transf,Y_transf,X_transf, alpha=0.25, edgecolors = 'red', label='$<2x,\, 3y,\, x-2y>$')
surf._facecolors2d=surf._facecolors3d 
surf._edgecolors2d=surf._edgecolors3d
ax1.legend()
ax1.set_xlabel("$x$")
ax1.set_ylabel("$y$")
ax1.set_zlabel("$z$")

ax1.view_init(20,60)

```
