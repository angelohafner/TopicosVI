```
import sympy as sp
import matplotlib.pyplot as plt
import numpy as np
```

```
t = sp.symbols('t')
x = sp.Function('x')
v = sp.diff(x(t), t)
a = sp.diff(v, t)
```

```
m = 1
b = 5
tt = np.linspace(0, 5, 1000)
fig = plt.figure(figsize=(10,5))
for k in range(1, 16, 2):
    edo = sp.Eq(m*a + b*v + k*x(t), 12)
    x_sol = sp.dsolve(edo, ics={x(0): 0, v.subs(t, 0): 0})
    sol_simbolica = x_sol.rhs
    sol_numerica = sp.lambdify(t, sol_simbolica)
    xx = sol_numerica(tt)
    plt.plot(tt, xx, label='k='+str(k))
plt.title('m='+str(m)+'  b='+str(b))
plt.legend(loc='best')
plt.grid()
plt.show()
```
