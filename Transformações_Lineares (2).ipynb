# Transformações Lineares


```python
import numpy as np
import matplotlib.pyplot as plt
```

Vetores unitários $\hat i$ e $\hat j$.

```python
vec_i = [1,0]
vec_i = np.array(vec_i)
vec_j = [0,1]
vec_j = np.array(vec_j)
lim = 4
```

Uma combinação linear 

$$\mathbf{v} = x \, \hat i + y \, \hat j$$

```python
vec_comb1 = -1*vec_i -3*vec_j
vec_comb1
```

## Transformação Linear

$${\rm{trLin}} = \left[ {\begin{array}{*{20}{c}}5&1\\3&3\end{array}} \right]$$


```python
trLin = [[5,1],[3,3]]
trLin = np.array(trLin) / 5
np.linalg.det(trLin)
```

Transformando os vetores $\hat i$, $\hat j$ e $\mathbf{v}$.

```python
vec_itr = np.matmul( trLin , vec_i )
vec_jtr = np.matmul( trLin , vec_j )
vec_comb1tr = np.matmul( trLin , vec_comb1 )
print('T(i) = ',vec_itr)
print('T(j) = ',vec_jtr)
print('T(v) = ',vec_comb1tr)
```

### AutoValores e AutoVetores

```python
aval,avec = np.linalg.eig(trLin)
print("AutoValores = ",aval)
print("\n")
print("AutoVetores = \n",avec)
```

Necessário para refazer as linhas de grade:

```python
x1 = avec[0,0]
y1 = avec[1,0]
ang2 = np.arctan(y1/x1) + np.pi/2
m2 = np.tan(ang2)
xx = np.linspace(-10*lim,10*lim)

x2 = avec[0,1]
y2 = avec[1,1]
ang3 = np.arctan(y2/x2) + np.pi/2
m3 = np.tan(ang3)
```

```python

fig, (ax1, ax2) = plt.subplots(1, 2,figsize=(12,5),dpi=150)


ax1.set_xlim(-lim,lim)
ax1.set_ylim(-lim,lim)
ax1.grid(axis='x',alpha=0.5, color='red', ls = '--')
ax1.grid(axis='y',alpha=0.5, color='blue', ls = '--')
ax1.quiver(0,0, vec_i[0],vec_i[1],        angles='xy', scale_units='xy', scale=1, color='red',                                 width = 0.005,  label='${\hat{i}}$')
ax1.quiver(0,0, vec_j[0],vec_j[1],        angles='xy', scale_units='xy', scale=1, color='blue',                                width = 0.005,  label='${\hat{j}}$')
ax1.quiver(0,0,vec_comb1[0],vec_comb1[1], angles='xy', scale_units='xy', scale=1, color='black',  headlength=9, headwidth = 5,                 label='${\mathbf{v}}$')

##########################################################################################################################

ax2.set_xlim(-lim,lim)
ax2.set_ylim(-lim,lim)
k = -20*y1
while( k < 20*y1 ):
    yy = m2*xx + k
    ax2.plot(xx,yy, color='red', lw=0.5, alpha=0.5, ls='--')
    k = k + y1

k = -20*y2
while( k < 20*y2 ):
    yy = m3*xx + k
    ax2.plot(xx,yy, color='blue', lw=0.5, alpha=0.5, ls='--')
    k = k + y2


ax2.quiver(0,0, vec_itr[0],vec_itr[1], angles='xy', scale_units='xy', scale=1, color='red',  width = 0.005, label='${\hat{i}}$')
ax2.quiver(0,0, vec_jtr[0],vec_jtr[1], angles='xy', scale_units='xy', scale=1, color='blue', width = 0.005, label='${\hat{j}}$')

avec = lim*avec
ax2.quiver(0,0, avec[0,0],avec[1,0], angles='xy', scale_units='xy', scale=1, color='red',  width = 0.01, headlength=9, headwidth = 5,  label='${\hat{v}_1}$')
ax2.quiver(0,0, avec[0,1],avec[1,1], angles='xy', scale_units='xy', scale=1, color='blue', width = 0.01, headlength=9, headwidth = 5,  label='${\hat{v}_2}$')

ax2.quiver(0,0,vec_comb1tr[0],vec_comb1tr[1], angles='xy', scale_units='xy', scale=1, color='black', headlength=9, headwidth = 5, label='${\mathbf{v}}$')

ax2.set_xlim(-lim,lim)
ax2.set_ylim(-lim,lim)

ax2.legend(loc='lower right')
ax2.grid(alpha=0.5)
```

```python
avec = avec/lim
lim = 1
```

# Imaginemos para várias entradas

```python
u = np.arange(-20*lim,21*lim,1)
v = u
n = len(v)
vec = np.zeros((2,n**2))
print(vec.shape)
j = 0
k = 0

for i in range(n):
    for j in range(n):
        vec[0,j+k] = u[i]
        vec[1,j+k] = v[j]
    k = k + n
vec
```

```python
vecTransf = np.matmul(trLin , vec)
vecTransf
```

```python
avecTransf = np.matmul(trLin , avec)
avecTransf
```

```python
aleatorios = [[0,-1],[2,0]]
aleatorios = np.array(aleatorios)
aleatoriosTransf = np.matmul( trLin , aleatorios )
aleatoriosTransf
```
