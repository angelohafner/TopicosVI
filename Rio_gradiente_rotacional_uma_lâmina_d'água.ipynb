```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,4,10)
y = np.arange(0,6,5)
print("x = ",x)
print("y = ",y)
```

```python
X,Y = np.meshgrid(x,y)
V = -X * ( X - 4 )
print("X = \n",X)
print("\nY = \n",Y)
print("\nV = \n",V)

Vy = V
Vx = 0*Vy
```

```python
plt.quiver(X,Y,Vx,Vy, scale_units='xy', angles='xy', scale = 1)
plt.ylim(0,10)
plt.grid()
```

```python
dv_dy, dv_dx = np.gradient(Vy,y,x)
```

```python
rotZ = dv_dx
rotZ.shape
```

```python
fig = plt.figure(figsize=(8,6),dpi=150)
ax1 = plt.axes(projection='3d')
ax1.quiver(X,Y,0*X, Vx,Vy,0*Vx, label='$v(x,y)$')
ax1.scatter(X,Y,0*X)
ax1.set_zlim(-5,5)

ax1.quiver(X,Y,0*X, rotZ*0,rotZ*0,rotZ, colors='red', label='$ rotacional $')
ax1.legend()
```
