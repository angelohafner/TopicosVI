```python
import numpy as np
```

```python
mTransicao = np.matrix([[0.7,0.15,0.15],[0.2,0.8,0.15],[0.1,0.05,0.7]])
print(mTransicao)
```

```python
vInicial = np.matrix([[0.2325],[0.2875],[0.48]])
print(vInicial)
```

```python
vProximo = np.matmul(mTransicao,vInicial)
```

```python
vProximo
```

```python
for k in range(1,16,1):
  vProximo = np.matmul(mTransicao,vInicial)
  vInicial = vProximo
  print(vProximo)
  print("\n\n")
```
