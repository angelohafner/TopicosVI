```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use(['science','notebook','grid'])
```

```python

def X(t):
    eq1 = np.exp(-t) + 3*np.exp(4*t)
    eq2 = -np.exp(-t) + 2*np.exp(4*t)
    return [eq1,eq2]
```

```python
t = np.linspace(-3,3,100)
```

```python
XX = X(t)
```

```python
plt.plot(t,XX[0],label=r'${e^{ - t}} + 3{e^{4t}}$')
plt.plot(t,XX[1],label=r'$ - {e^{ - t}} + 2{e^{4t}}$')
plt.legend()
plt.xlabel('$t$')
plt.ylim([-6,6])
plt.show()
```

```python
plt.xlim([0,15])
plt.ylim([-10,5])
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.plot(XX[0],XX[1])
plt.show()
```
