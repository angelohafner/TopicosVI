### Resolva
$$(y\cos x + 2x{e^y}) + (\sin x + {x^2}{e^y} - 1)y' = 0$$

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
```

```python
y = sp.Function("y")
x = sp.Symbol('x')
y(x)
```

```python
D1 = sp.Derivative(y(x), x)
D1
```

```python
M = y(x) * sp.cos(x) + 2*x*sp.exp(y(x))
M
```

```python
N = sp.sin(x) + x**2 *sp.exp(y(x)) - 1
N
```

```python
ode = sp.Eq(M + N*D1,0)
ode
```

Não tem o que resolva, por isto foi resolvido na mão e deixado a solução na forma implícita
$$y\sin x + {x^2}{e^y} - y = c$$

```python
xx = np.linspace(0,2,25)
yy = np.linspace(0,6,25)
XX,YY = np.meshgrid(xx,yy)
```

```python
num = YY * np.cos(XX) + 2*XX*np.exp(YY)
den = np.sin(XX) + XX**2 * np.exp(YY) - 1
deriv = -num/den
ang = np.arctan(deriv)
u = np.cos(ang)
v = np.sin(ang)

fig = plt.figure(dpi=300,figsize=(8,4))

plt.quiver(XX,YY,u,v,angles='xy')
xx = np.linspace(0,2,170)
yy = np.linspace(0,6,170)
XX,YY = np.meshgrid(xx,yy)
for cc in range(-8,9,2):
  ZZ = YY * np.sin(XX) + XX**2*np.exp(YY) - YY - cc
  plt.contour(XX,YY,ZZ,[0],label=cc)


plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()
```
