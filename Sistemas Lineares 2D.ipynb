```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
aa = 10
x = np.linspace(-aa,aa,100)
```

```python
A = np.matrix([[2,-1],
               [-3,-1]])
A
```

```python
b = np.matrix([[3],
               [7]])
b
```

```python
xSol = np.linalg.solve(A,b)
xSol = np.array(xSol)
xxx = xSol[0,0]
yyy = xSol[1,0]
print(xxx)
print(yyy)
```

```python
#print(plt.style.available)
A[[1],[1]]
b[[1],[0]]
```

```python
x = np.linspace(-4,4,100)

y1 = ( -A[[0],[0]]*x + b[[0],[0]] ) / A[[0],[1]]
y2 = ( -A[[1],[0]]*x + b[[1],[0]] ) / A[[1],[1]]

plt.figure(dpi=150)
plt.plot(x,y1.transpose(),label='$-2x-3$')
plt.plot(x,y2.transpose(),label='$-3x-7$')
plt.scatter(xxx,yyy,s = 80,facecolors='none', edgecolors='r',label=xSol)
plt.grid()
plt.title('Interpretação por linhas de um Sistema Linear 2D')
plt.legend()
```

```python
plt.plot(x,y1,label='$y=2x+3$')
plt.plot(x,y2,label='$y=-3x+7$')
plt.grid()
plt.xlabel('$x$')
plt.xlabel('$y$')
plt.legend()
plt.show()
```
