# Gráfico de Funções Implícitas
$${y^2} - 2y = {x^3} + 2{x^2} + 2x + 3$$

```python
import matplotlib.pyplot plt
import numpy as np

```

```python
delta = 0.025
xrange = np.arange(-5.0, 20.0, delta)
yrange = np.arange(-5.0, 20.0, delta)
X, Y = np.meshgrid(xrange,yrange)
```

$F$ é a função de um lado e $G$ de outro lado.
O nível zero do `contour` é para plotar onde as duas funções são iguais.

```python
F = Y**2 - 2*Y
G = X**3 + 2*X**2 +2*X + 3
plt.contour(X, Y, (F - G), [0])
plt.show()
```
