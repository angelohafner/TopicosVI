```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-2,2,20)
y = np.linspace(-2,2,20)
```

```python
X,Y = np.meshgrid(x,y)
plt.scatter(X,Y)
```

```python
Z = X**2 - Y**2
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z)
```
