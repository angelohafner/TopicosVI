```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def paraboloide(x,y):
    saida = ((-x**2-y**2) + 4)
    return [saida]
```

```python
dx = 0.2
x = np.arange(-3,3.1,dx)
```

```python
z = np.arange(-2,5.1,dx)
xPlan,zPlan = np.meshgrid(x,z)
b0 = 0.5
yPlan = b0*xPlan/xPlan
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(xPlan,yPlan,zPlan)
```

# Superfcie Gerada

$$ z(x,y) = -x^2 -y^2 + 4 $$

```python
y = np.arange(-3,b0+0.1,dx)
[xSurf,ySurf] = np.meshgrid(x,y,indexing='ij')
zSurf = (-xSurf**2-ySurf**2) + 4
#zSurf[zSurf<-2] = np.nan
```

# Inteseco com o plano $y=0,5$

$${\rm Intersec}(x,y,z) = < x\, ; \; 0,5 \, ; \; -x^2-0,5^2+4 >$$

```python
xIntersec = x
yIntersec = 0.5*x/x
zIntersec = np.array((-xIntersec**2-yIntersec**2) + 4)
zIntersec[zIntersec<-2] = np.nan
```

# Ponto de $y=2$ onde se deseja saber a derivada parcial $z_x$

$x_P = 1\\
  y_P = 0,5\\
  z_P = -(1)^2 - (0,5)^2 + 4 $

```python
xP = 1
yP = 0.5
zP = -xP**2-yP**2 + 4
```

# Equao da reta tangente no ponto $P$

A equao de qualquer reta  dada por $z - {z_0} = m\left( {x - {x_0}} \right)$.

Aqui ${z_x} =  - 2x$ e a equao da reta fica:

$$z = {z_P} +  - 2{x_P}\left( {x - {x_P}} \right)$$



```python
xR= x
zR = zP - 2*xP*(xR-xP)
yR = 0.5*x/x
zR[zR>6] = np.nan
zR[zR<-2] = np.nan
```

# Finalmente, os grficos

```python
fig = plt.figure(dpi=200,figsize=(16,6))
ax = plt.axes(projection='3d')
ax.plot_surface(xSurf,ySurf,zSurf,alpha=0.3,color='green')
ax.contour3D(xSurf,ySurf,zSurf,colors='green',linestyles ='dashed')
ax.plot_surface(xPlan,yPlan,zPlan,alpha=0.2)
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
ax.set_zlabel('$z$')
ax.set_zlim(0,4)
ax.plot(xIntersec,yIntersec,zIntersec,lw=5,  label = "interseco", color='green')
ax.scatter(xP,yP,zP,color='blue', label = "ponto tangente")
ax.plot(xR,yR,zR,lw=2, ls='--', label = "reta tangente", color='red')

ax.quiver(xP,yP,zP, -2*xP, 0,    0, label="$\partial f / \partial x$", color='red')
ax.quiver(xP,yP,zP,  0,   -2*yP, 0, label="$\partial f / \partial y$", color='blue')
plt.title("Derivada parcial com relao a $x$")
ax.legend()
ax.view_init(20,110)
fig.show()
```
