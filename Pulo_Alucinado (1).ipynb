```python
import numpy as np
import matplotlib.pyplot as plt
```

$$ v'=10-\frac{1}{6}v$$

```python
tempo = np.linspace(0, 20, 11)
veloc = np.arange(-10, 90, 10)
```

```python
TT, VV = np.meshgrid(tempo, veloc)
inclinacoes = 10 - 1/6 * VV
angulos = np.arctan(inclinacoes)
compX = np.cos(angulos)
compY = np.sin(angulos)
```

```python
plt.quiver(TT, VV, compX, compY, angles='xy')
for gustavo in range(-80, 30, 10):
    veloc_2 = gustavo*np.exp(-tempo/6)+60
    plt.plot(tempo, veloc_2)
plt.show()
```
